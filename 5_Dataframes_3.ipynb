```python
def reimport():
    df = pd.read_csv('jamesbond.csv', index_col='Film')
    df.sort_index(inplace=True)
    return df
```

# Second arguments to `.loc[], .iloc[]` methods

```python
df = reimport()
df.set_index('Film', inplace=True)
df.sort_index(inplace=True)
df.head(3)
```

```python
df.loc['Moonraker']
```

```python
# if you just want to pull the director at that row
df.loc['Moonraker','Director']
```

```python
# or if you want to pull Director through Budget
df.loc['Moonraker', 'Director':'Budget']
```

```python
# or if you want specific columns
df.loc['Moonraker', ['Actor','Budget','Year']]
```

```python
# the same things is true with the .iloc method
df.iloc[1,[0,2,3]] # note that '0' is year, not the file name
```

# Review of the `.apply` method on single columns

```python
df = reimport()
df.head()
```

```python
def convert_to_string_add_millions(number):
    return str(number) + ' MILLIONS!'
```

```python
df['Box Office'] = df['Box Office'].apply(convert_to_string_add_millions)
```

```python
df.head()
```

```python
columns = ['Box Office', 'Budget', 'Bond Actor Salary']
for col in columns:
    df[col] = df[col].apply(convert_to_string_add_millions)
df.head()
```

# the `.apply` method on row values

```python
df = reimport()
df.head()
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    if actor == 'Pierce Brosnan':
        return 'the best'
    elif actor == 'Roger Moore' and budget > 40:
        return 'pretty good'
    else:
        return 'who knows'

df.apply(good_movie, axis='columns') # this is really confusing axis=columns
```

# the `.copy()` method

```python
df = reimport()
df.head()
```

```python
directors = df['Director']
directors
```

```python
directors['A View to a Kill']
```

```python
directors['A View to a Kill'] = 'Mr. ' + directors['A View to a Kill']
```

```python
directors
```

```python
# note the original value was affected 
df.head(3)
```

```python
df = reimport()
directors = df['Director'].copy()
```

```python
directors['A View to a Kill'] = 'Mr. ' + directors['A View to a Kill']
```

```python
df.head() # now unaffected because the operation was done on the copy
```
