```python
import pandas as pd
```

# Feed `pd.read_csv()` method a url argument

```python
# https://data.cityofnewyork.us/Health/Most-Popular-Baby-Names-by-Sex-and-Mother-s-Ethnic/25th-nujf
def baby_names():
    url = 'https://data.cityofnewyork.us/resource/g374-eanh.csv'
    return pd.read_csv(url)
```

# Quick object conversions

```python
names = baby_names()
```

```python
names.head()
```

## to python list

```python

names['nm'].tolist()
```

## to dataframe

```python
#converts a column to a one-column frame
names['nm'].to_frame()
```

## returning a string of comma-separated values

```python

", ".join(str(name) for name in names['nm'])
```

# Export CSV with the `to_csv()` method

```python
names = baby_names()
```

```python
names.to_csv('NYCBabyNames.csv', index=False, columns=['nm','brth_yr'], encoding='utf-8')
```

```python
out = pd.read_csv('NYCBabyNames.csv')
out.head()
```

# Export Excel File

```python
names = baby_names()
names
```

```python
abby = names[names['nm']=='ABIGAIL']
ada = names[names['nm']=='ADA']
```

```python
# create an excel writer object
my_file = pd.ExcelWriter('BabyNames.xlsx')
```

```python
# write to the object, adding sheet names, etc.
abby.to_excel(my_file, sheet_name='Abby', index=False)
ada.to_excel(my_file, sheet_name='Ada', index=False)
```

```python
# save the file
my_file.save()
```
